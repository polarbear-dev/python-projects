```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np


%matplotlib inline 
```

The *avocado_mean* contains aggregated data

```python
avocado_mean = pd.read_csv("https://stepik.org/media/attachments/lesson/362453/avocado_mean.csv",
                           index_col=0, parse_dates=['Date'])
```

```python
avocado_mean.head()
```

Calculate the rolling average of avocado prices (*AveragePrice*) with a *window* of 3 and find maximum value

```python
rm = avocado_mean.rolling(window=3).mean()
rm.head()
```

```python
np.round(rm.max(), 2)
```

See how the graph changes depending on the selected window size.

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 10), sharey='col', sharex=True)

windows = [2, 4, 10, 50]
colors = ['coral', 'blue', 'green', 'purple']

for window, ax, color in zip(windows, axes.flatten(), colors): 
    ax.plot(avocado_mean.rolling(window=window).mean(), label=window, color=color)

for ax in axes.flatten():
    # remove the frame
    ax.set_frame_on(False)  
    # set major locator – January 4th for each year
    ax.xaxis.set_major_locator(mdates.MonthLocator(bymonth=1, bymonthday=4))  
    # show in the format of abbreviated month name and day (Jan 04)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
    # under the major locator - minor locator, i.e., edit minor ticks
    ax.xaxis.set_minor_locator(mdates.YearLocator(month=1, day=1))
    # show the year
    ax.xaxis.set_minor_formatter(mdates.DateFormatter('\n%Y'))
    # make the grid lines slightly gray and half transparent
    ax.grid(True, color='#e2e2e2', alpha=0.5)

for name, ax in zip(['1','2','3','4'], axes.flatten()):
    ax.set(title='Chart {}'.format(name))
    ax.tick_params(labelbottom=True, which='both')
```

Apply the *ewm()* function with the parameter *span=2* to obtain the exponential moving average

```python
avocado_ewm = avocado_mean.ewm(span=2).mean()
avocado_ewm.head()
```

Import the complete data

```python
avocado_full = pd.read_csv("https://stepik.org/media/attachments/lesson/362453/avocado_full.csv",
                           index_col=0, parse_dates=['Date'])
```

```python
avocado_full.head()
```

For avocados of *type* organic in the Chicago *region*, calculate the rolling average with a window of 4 and the exponential moving average with a parameter *span=4*

```python
organic_ch = avocado_full.loc[(avocado_full.type == 'organic') & (avocado_full.region == 'Chicago')]
```

```python
fig, ax = plt.subplots(figsize=(16,4))
organic_ch.AveragePrice.rolling(window=4).mean().plot(ax=ax, label='rolling')
organic_ch.AveragePrice.ewm(span=4).mean().plot(ax=ax, label='ewm')
plt.legend()
plt.grid()
```
